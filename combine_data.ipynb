```python
import pandas as pd
```

```python
mean_income = pd.read_csv("mean-income-or-consumption-per-day_OURWORLDINDATA.csv")

mean_income = mean_income.drop(columns=["Year"])

mean_income = mean_income.rename(columns={"Mean income or consumption per day": "Income Per Day"})

mean_income.head()
```

```python
gdp = pd.read_csv("GDP_per_country_world_bank.csv")

# Rename column so both have 'Country'
gdp = gdp.rename(columns={"Data Source": "Country"})

common_countries = set(mean_income["Country"]).intersection(set(gdp["Country"]))

gdp = gdp[gdp["Country"].isin(common_countries)]
mean_income = mean_income[mean_income["Country"].isin(common_countries)]


gdp = gdp.sort_values("Country").reset_index(drop=True)
mean_income = mean_income.sort_values("Country").reset_index(drop=True)



gdp = gdp[["Country", "Unnamed: 68"]]

# Rename the column
gdp = gdp.rename(columns={"Unnamed: 68": "GDP in USD"})

gdp.head(10)
```

```python
hdi = pd.read_excel("HDR25_Statistical_Annex_HDI_Table.xlsx")

# Drop rows 0 through 6
hdi = hdi.drop(index=range(7))


hdi = hdi.rename(columns={
    "Table 1. Human Development Index and its components": "Country",
    "Unnamed: 2": "HDI Value"
})

hdi = hdi[["Country", "HDI Value"]]

common_countries = set(mean_income["Country"]).intersection(set(hdi["Country"]))


hdi = hdi[hdi["Country"].isin(common_countries)]
mean_income = mean_income[mean_income["Country"].isin(common_countries)]


hdi = hdi.sort_values("Country").reset_index(drop=True)
mean_income = mean_income.sort_values("Country").reset_index(drop=True)

hdi.head(20)
```

```python
population = pd.read_excel("pop_per_country_world_bank.xls")

population = population.drop(index=range(3))

# Rename columns
population = population.rename(columns={
    "Data Source": "Country",
    "Unnamed: 68": "Population"
})

# Keep only Country and Population columns
population = population[["Country", "Population"]]

# Strip whitespace from Country names
population["Country"] = population["Country"].str.strip()
mean_income["Country"] = mean_income["Country"].str.strip()

common_countries = set(mean_income["Country"]).intersection(set(population["Country"]))


population = population[population["Country"].isin(common_countries)]
mean_income = mean_income[mean_income["Country"].isin(common_countries)]


population = population.sort_values("Country").reset_index(drop=True)
mean_income = mean_income.sort_values("Country").reset_index(drop=True)

population.head(10)
```

```python
combined = pd.concat([
    hdi[['Country']],        # only keep the Country column once
    hdi.drop(columns='Country'),
    gdp.drop(columns='Country'),
    mean_income.drop(columns='Country'),
    population.drop(columns='Country')
], axis=1)

combined.head(20)
```

```python
combined_clean = combined.dropna()
combined_clean = combined_clean.iloc[:146]
combined_clean.to_csv('country_data_combined.csv', index=False)
```
